```python
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder
```

```python
train_df = pd.read_csv(Path('Resources/2019loans.csv'))
test_df = pd.read_csv(Path('Resources/2020Q1loans.csv'))
```

```python
#review train_df of new generates CSV
train_df.head()
```

```python
#review test_df
test_df.head()
```

```python
# Convert categorical data to numeric and separate target feature for training data
#convert data using get.dummies
#Drop the taget column
X = train_df.drop(['target'], axis=1)
X
```

```python
#convert data using get.dummies
X_train = pd.get_dummies(X)
print(X_train.columns)
X_train.head()
```

```python
# Converting training output labels to 0 and 1 using LabelEncoder
y_train = LabelEncoder().fit_transform(train_df['target'])
y_train
```

```python
# Convert categorical data to numeric and separate target feature for testing data
X = test_df.drop(['target'], axis=1)

X
```

```python
#convert data using get.dummies
X_test = pd.get_dummies(X)
print(X_test.columns)
X_test.head()
```

```python
#Find missing test column
set(X_train.columns) ^ set(X_test.columns)
```

```python
# add missing dummy variables to testing set
X_test["debt_settlement_flag_Y"] = 0

X_test.head()
```

```python
# Converting testing output labels to 0 and 1 using LabelEncoder
y_test = LabelEncoder().fit_transform(test_df['target'])
y_test
```

# Model Prediction
My prediction is that the Random Forest Classifier model will perform better. I believe this will be true for the following reasons:
 - There are a number of categorical columns in the dataset and a Random Forest Classifier may perform better
 - There are a number of variables in the dataset and a Random Forest Classifier may perform better and capture feature patterns
 - With Random Forest Classifier you could also feature selection to select only the relevant values in the dataset 

```python
# Train the Logistic Regression model on the unscaled data and print the model score
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(solver="liblinear", max_iter=200).fit(X_train,y_train)

print("Testing Score:", lr.score(X_test, y_test))
print("Training Score:", lr.score(X_train, y_train))
```

```python
# Train a Random Forest Classifier model and print the model score
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(random_state=13, n_estimators=300).fit(X_train, y_train)

print(f'Testing Score: {clf.score(X_test, y_test)}')
print(f'Training Score: {clf.score(X_train, y_train)}')
```

# Impact of Scaling
My assumption is that scaling the data will improve the accuracy and performance of both models for the following reasons:
- The Logistic Regression model threw an alert message stating that the total number of iterations reached the max limit and that the data needs to be scaled. I resolved this alert by increasing the max_iter, but the alert would also be resolved after scaling the data and scaling will improve the performance of the Logistic Regression model.
- Reducing the magnitude of outliers or large values that can impact the accuracy of the models due the large size will improve performance.
- There is a large range of values in the dataset. The machine learning algorithm does not know what the number represents and could make erroneous assumptions if not scaled.


```python
# Scale the data
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Train the Logistic Regression model on the scaled data and print the model score
lrs = LogisticRegression(solver="liblinear", max_iter=200)

lrs.fit(X_train_scaled, y_train)
print("Testing Score:", lrs.score(X_test_scaled, y_test))
print("Training Score:", lrs.score(X_train_scaled, y_train))
```

```python
# Train a Random Forest Classifier model on the scaled data and print the model score
clfs = RandomForestClassifier(random_state=13, n_estimators=300, max_depth=3).fit(X_train_scaled, y_train)
print(f'Testing Score: {clfs.score(X_test_scaled, y_test)}')
print(f'Training Score: {clfs.score(X_train_scaled, y_train)}')
```

# Model Comparisons

## Performance
Prior to scaling, the models perform similarly with an edge to Random Forest. However, with Random Forest the accuracy is at 64% for the testing score and is 100% for the training score, which raises concerns of over-fitting.

## Scaling Improvements
Scaling the data had a significant improvement on the Testing Score of the Logistic Regression model and resolved the alert message, but scaling did not seem to have much of an impact on the Random Forest Classifier. Based on Google research, Random Forest does not necessarily need scaling since the nature of the model does not present the numerical precision issues found with Logistic Regression.

## Final Thoughts
I am going to change my position on the models and give the Logistic Regression an edge over Random Forest. After scaling the Logistic Regression model performs better than Random Forest with a testing score of 76 % over the Random Forest testing score of 62%. The Random Forest model had a perfect training score at 100%, this would indicate that the model is over-fitting. I limited the depth to 3, which seemed to help, but the model still appears to over-fit. Perhaps if feature selection was applied the chances of over-fitting could be reduced, but overall it appears that the Logistic Regression model peforms better with this dataset.
